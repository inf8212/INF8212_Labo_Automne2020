### 1.	Choisir aléatoirement un élément d’une liste et l’afficher.

Utiliser d’abord la fonction random() du module random; essayer ensuite avec la fonction randrange() du même module.

Vous pouvez aller voir [ici](https://www.w3schools.com/python/module_random.asp) pour des exemples.

```python
# Soit une liste quelconque...
participants = [
  'exemple1',
  'exemple2',
  'exemple3',
  'exemple4'
]

# La fonction random() retourne un réel entre 0.0 et 1.0
# On utilise cette fraction pour générer un entier correspondant à un des index de la liste (ici on veut 0, 1, 2 ou 3)
# On sélectionne un élément de la liste avec cet index généré par le hasard (pseudo-random)
############################################
from random import random

fraction    = random()
index       = int(len(participants)*fraction) # int() va tronquer les fractions
happyChosen = participants[index]
print(happyChosen)


# La fonction randrange() retourne un entier entre 0 et l'entier passé en paramètre -1.
# Ici on passe la longeur de la liste comme paramètre de fin.
# On peut aussi passer un paramètre de départ au besoin
############################################
from random import randrange

index       = randrange(len(participants))
happyChosen = participants[index]
print(happyChosen)
```

### 3. Affecter des bourses à des étudiants selon une liste de critères

Soit une liste d’étudiant : chaque élément de la liste correspond à un étudiant; cet élément contient lui-même une liste dont le 1er élément est le code permanent et le second élément est (encore!) une liste; chaque élément de cette dernière liste est une bourse pour laquelle l’étudiant pourrait se qualifier.

Retournez voir la définition du code permanent ici : https://etudier.uqam.ca/code-permanent-uqam

<pre><code>
# Voici la liste des étudiants avant l’affectation des bourses
# Notez la liste vide [] comme 2e élément de chaque élément de la liste. On va y ajouter les bourses
etudiants = [
    ['ABCD68040102', []],
    ['EFGH15069501', []],
    ['IJKL27599304', []]
]

# Voici la liste des bourses (le nom donne le critère…)
bourses = [
    'Bourse pour domaines non-traditionnels - femmes',
    'Bourse pour domaines non-traditionnels - hommes',
    'Bourse pour les jeunes (nés en 2000 ou après)'
]
</code></pre>

**Votre mission : affecter les bourses aux bons étudiants!**


```python
# Variables
#################################################

etudiants = [
    ['ABCD68040102', []],
    ['EFGH15069501', []],
    ['IJKL27599304', []]
]

bourses = [
    'Bourse pour domaines non-traditionnels - femmes',
    'Bourse pour domaines non-traditionnels - hommes',
    'Bourse pour les jeunes (nés en 2000 ou après)'
]

# Pour qu'on se comprenne bien avec la liste etudiants:
# etudiants renvoie toute la liste
# etudiants[0] renvoie le 1er élément de la liste, correspondant au 1er étudiant
# etudiants[0][0] correspond au code permanent du 1er étudiant
# etudiants[0][1] correspond à la liste (vide pour commencer) des bourses du 1er étudiant

# En itérant sur la liste étudiants avec 'for etudiant in etudiants',
# la variable etudiant (sans le s à la fin!) correspond à un étudiant de la liste à chaque itération
# On a donc etudiant qui renvoie une liste contenant le code permanent et la liste des bourses
# On a etudiant[0] qui renvoie le code permanent (une chaîne)
# On a etudiant[1] qui renvoie la liste des bourses (une liste)


# Affecter les bourses...
# On convertie les tranches de chaine en entier avant de faire les comparaisons arithmétiques.
# On utilise la méthode append() pour ajouter un élément à la fin d'une liste
#################################################

for etudiant in etudiants:
    if int(etudiant[0][6:8]) > 50: etudiant[1].append(bourses[0]) # Femme avec mois de naissance > 50
    else                         : etudiant[1].append(bourses[1]) # Homme avec mois de naissance <= 50 (entre 01 et 12 en fait)
    if int(etudiant[0][4:6]) > 62: etudiant[1].append(bourses[2]) # Nés en 2000 ou après avec jour de naissance > 62

# On affiche tout ça de façon un peu élégante!
for etudiant in etudiants:
    print(etudiant[0], ':')
    for bourse in etudiant[1]: print('\t', bourse)
```

### 4. Vérifier si une chaîne de caractère est un palindrome.

On va faire l’exercice si ça n’a pas déjà été fait en classe avec Alix. Il y a plusieurs façon élégantes de procéder; on va en voir deux. On va aussi essayer d’optimiser notre programme pour qu’il effectue la tâche le plus rapidement possible.

```python
# D'abord, demandons à l'utilisateur pour une expression
# Notez l'utilisation de la méthode replace() pour enlever les espaces (On ne veut pas les considérer pour notre palindrome)
# Notez l'utilisation de la méthode upper() pour tout mettre en majuscule; 
# on aurait aussi pu mettre tout en minuscule, l'important c'est de s'assurer que 'a'=='A' par exemple
# Voyez comment les fonctions peuvent s'enchaîner: input() retourne une chaine,
# replace() prend cette chaine et en retourne une autre, idem pour upper()


a = input("Palindrome svp (Ex. 'Karine alla en irak') : ").replace(' ', '').upper()


# 1ere façon
# On fait un itération sur la longueur de la chaîne et on compare la valeur du premier caractère avec le dernier
# Si pareils, on compare le second avec l'avant dernier, et ainsi de suite
#######################################
result = True # On commence par supposer que la chaîne est un palindrome, et on tente de prouver l'inverse
for i in range(len(a)):
    if a[i] != a[-i-1]:
        result = False
        break # Permet de sortir prématurément de la boucle
print(result)


# Même chose avec une boucle while
#######################################
result, i = True, 0 # La variable i va être utilisée pour faire référence à l'index
while result and i < len(a):
    if a[i] != a[-i-1]: result = False
    i+=1 # incrémente i de +1. Équivalent à i = i + 1
print(result)


        
# 1ere façon optimisée
# Pas besoin d'itérer sur toute la chaîne; à partir du milieu on a fini en fait.
# Le +1 permet de gérer les nombres impairs de caractères.
#######################################
result = True
for i in range(len(a)+1//2):
    if a[i] != a[-i-1]:
        result = False
        break
print(result)



# 2eme façon : On compare la chaine avec son inverse...
# Les listes ont une méthode reverse() bien utile, donc on transforme d'abord en liste
# puis on renverse la liste, puis on re-transforme en chaîne
# en concaténant (join) les éléments de la liste avec un séparateur vide
# Attention, la méthode reverse() effectue une opération sur une liste existante, et retourne toujours None
# On doit donc faire ces opérations sur des lignes séparées.
#######################################
a_lst = list(a)
a_lst.reverse()
print(a==''.join( a_lst ))



# 2eme façon, mais plus élégant?
# on accède à des tranches (slices) de chaine avec [star:stop:step]
# :step n'est pas obligatoire mais ici il est bien utile! En faisant des pas de -1, on itère sur la chaine dasn l'autre sens!
# a[:] donne toute la chaine; a[::-1] donne toute la chaine inversée
#######################################
print(a==a[::-1])

```
